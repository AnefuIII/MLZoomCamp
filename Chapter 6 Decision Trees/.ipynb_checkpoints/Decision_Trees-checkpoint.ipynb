```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CreditScoring.csv')
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.status.value_counts()
```

```python
df.status = df.status.map({1: 'ok', 2: 'default', 0: 'unk'})
```

```python
df.head()
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk',
}

df.home = df.home.map(home_values)


marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'seperated',
    5: 'divorced',
    0: 'unk',
}

df.marital = df.marital.map(marital_values)


record_values = {
    1: 'no',
    2: 'yes',
    0: 'unk',
}

df.records = df.records.map(record_values)

job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unk',
}

df.job = df.job.map(job_values)

```

```python
df.head(2)
```

```python
df.describe().round()
```

```python
# replace 9999999. they are missing values

df.income.replace(to_replace = 99999999.0, value = np.nan).max()
```

```python
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace = 99999999.0, value = np.nan)
```

```python
df.describe().round()
```

```python
df = df[df.status != 'unk'].reset_index(drop = True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size = 0.2, random_state = 11)

df_train, df_val =  train_test_split(df_full_train, test_size = 0.25, random_state = 11)

print('df_full_train length is %s' %len(df_full_train))
print('df_test length is %s' %len(df_test))
print('df_val length is %s' %len(df_val))
```

```python
df_train = df_train.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
df_val = df_test.reset_index(drop = True)
```

```python
df_train.head()
```

```python
(df_train.status == 'default').astype('int').values
```

```python
y_train = (df_train.status == 'default').astype('int').values
y_test = (df_test.status == 'default').astype('int').values
y_val = (df_val.status == 'default').astype('int').values
```

```python
del df_train['status']
del df_test['status']
del df_val['status']
```

```python
df_train.head()
```

### Training

```python
def assess_risk(client):
    if client['records'] == 'yes':
        if client['job'] == 'partime':
            return 'default'
        else:
            return 'ok'
    else:
        if client['assets'] > 6000:
            return 'ok'
        else:
            return 'default'
```

```python
df_train.iloc[0].to_dict()
```

```python
xi = df_train.iloc[0].to_dict()
```

```python
assess_risk(xi)
```

### Training the DecisiontTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dict = df_train.fillna(0).to_dict(orient = 'records')

dv = DictVectorizer(sparse = False)

X_train = dv.fit_transform(train_dict)
```

```python
dv.feature_names_
```

```python
X_train
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

### test

```python
val_dict = df_val.fillna(0).to_dict(orient = 'records')
X_val = dv.transform(val_dict)
```

```python
y_pred = dt.predict_proba(X_val)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_val, y_pred)
```

```python
train_pred = dt.predict_proba(X_train)[:,1]
roc_auc_score(y_train, train_pred)
```

#### it is overfitting therefore we reduce the depth of our tree. ATree that is to short is called a Decision stump

```python
dt = DecisionTreeClassifier(max_depth = 3)
dt.fit(X_train, y_train)
```

```python
train_pred = dt.predict_proba(X_train)[:,1]
auc = roc_auc_score(y_train, train_pred)
print('Train auc: ', auc)

y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print('Val auc: ', auc)
```
